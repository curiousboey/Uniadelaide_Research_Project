```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the Excel file
df = pd.read_excel('AI data synk.xlsx')

# Extract the relevant columns
applications = list(df['AI Application'])
severity_data = df['Severity (Critical, High, Medium, Low)'].str.strip('()').str.replace(' ', '').str.split(',')

# Convert the severity data into a list
severity_data_list = [[int(val) for val in severity] for severity in severity_data]

# Create a list of severity categories
severity_categories = ['Critical', 'High', 'Medium', 'Low']

# Define colors for severity categories
colors = ['#FF0000', '#FFA500', '#FFFF00', '#008000']

# Create a stacked bar plot for each severity category for each application
fig, ax = plt.subplots(figsize=(10, 6))

# Initialize a list to track the bottom values for the bars
bottom_values = [0] * len(applications)

# Loop through each severity category and plot a stacked bar for each application
for i, category in enumerate(severity_categories):
    severity_values = [severity[i] for severity in severity_data_list]
    
    # Add black borders to each bar
    ax.bar(applications, severity_values, label=category, color=colors[i], bottom=bottom_values, edgecolor='black')
    
    bottom_values = [bottom_values[j] + severity_values[j] for j in range(len(applications))]

# Set labels and title
ax.set_xlabel('AI Applications')
ax.set_ylabel('Severity Counts')
ax.set_title('Severity Distribution in AI Applications')

# Add a legend
ax.legend(title='Severity', loc='upper right', bbox_to_anchor=(1.2, 1))

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha="right")

# Add grid lines
ax.yaxis.grid(True, linestyle='--', alpha=0.7)

# Customize colors and styles for the grid lines
ax.set_axisbelow(True)
ax.yaxis.grid(color='gray', linestyle='dashed', linewidth=0.5)

# Adjust the bar width for a cleaner appearance
for container in ax.containers:
    container.width = 0.6

# Show the stacked bar plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the Excel file
df = pd.read_excel('AI data synk.xlsx', engine='openpyxl')

# Extract relevant data
applications = df['AI Application']
total_dependencies = df['Total Dependencies']
vulnerable_dependencies = df['Issues in Dependencies']

# Calculate the percentage of Issues in Dependencies relative to total dependencies
percentage_vulnerable = (vulnerable_dependencies / total_dependencies) * 100

# Create a figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Create a single stacked bar chart for each application with black borders
ax.bar(applications, total_dependencies, label='Total Dependencies', color='b', edgecolor='black', linewidth=1)
ax.bar(applications, vulnerable_dependencies, label='Issues in Dependencies', color='g', bottom=total_dependencies, edgecolor='black', linewidth=1)

# Add labels and title
ax.set_xlabel('AI Application')
ax.set_ylabel('Count')
ax.set_title('Percentage of Issues in Dependencies over Total Dependencies by Application')
ax.set_xticklabels(applications, rotation=45, ha='right')
ax.legend()

# Display the percentages as text on top of the bars
for app, percentage in zip(applications, percentage_vulnerable):
    ax.text(app, total_dependencies[applications == app].values[0] + 5, f'{percentage:.1f}%', ha='center')

# Add grid lines
ax.yaxis.grid(True, linestyle='--', alpha=0.7)

# Customize colors and styles for the grid lines
ax.set_axisbelow(True)
ax.yaxis.grid(color='gray', linestyle='dashed', linewidth=0.5)

# Show the stacked bar plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the Excel file
df = pd.read_excel('AI data synk.xlsx', engine='openpyxl')

# Extract relevant data
applications = df['AI Application']
total_dependencies = df['Total Dependencies']
vulnerable_dependencies = df['Issues in Dependencies']


# Calculate the percentage of Issues in Dependencies relative to total dependencies
percentage_vulnerable = (vulnerable_dependencies / total_dependencies) * 100

# Set the size and color of bubbles
bubble_size = vulnerable_dependencies * 20  # Adjust bubble size for better visualization
bubble_color = percentage_vulnerable

# Create labels for applications with percentage values
application_labels = [f'{app} ({percentage:.1f}%)' for app, percentage in zip(applications, percentage_vulnerable)]

# Create a visually appealing bubble plot
plt.figure(figsize=(12, 8))
plt.scatter(total_dependencies, application_labels, s=bubble_size, c=bubble_color, cmap='viridis', alpha=0.6, edgecolors='k', linewidths=1)

# Customize the AI Application labels
plt.ylabel('AI Application', rotation=0, ha='right', weight='bold', fontsize=12)
plt.xlabel('Total Dependencies', fontsize=12)
plt.title('Bubble Plot of Total Dependencies and Issues in Dependencies by Application', fontsize=14)

# Create a colorbar for the bubble color legend
cbar = plt.colorbar()
cbar.set_label('Percentage of Issues in Dependencies', fontsize=12)

# Adjust the y-axis labels for better spacing
plt.yticks(fontsize=10)

plt.grid(True, axis='x', linestyle='--', alpha=0.5)

# Show the visually appealing bubble plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the data from the Excel file
df = pd.read_excel('AI data synk.xlsx')

# Extract the relevant columns
severity_data = df[['AI Application', 'Severity (Critical, High, Medium, Low)']]

# Create a list of severity categories
severity_categories = ['Critical', 'High', 'Medium', 'Low']

# Create a radar chart to plot the severity data
fig = plt.figure(figsize=(10, 10))
ax = plt.subplot(111, polar=True)

# Define the number of categories (severity levels) and the angle at which each category is located
num_categories = len(severity_categories)
angles = np.linspace(0, 2 * np.pi, num_categories, endpoint=False).tolist()

# Set the radar chart's axis labels and ticks
ax.set_xticks(angles)
ax.set_xticklabels(['Critical', 'High', 'Medium', 'Low'])  # Replace with severity levels
plt.title('Severity Radar Chart')

# Plot concentric circles for grid lines
for i in range(1, 6):
    ax.plot(angles, [i] * num_categories, linestyle='dashed', color='gray')

# Loop through each row in the dataframe
for index, row in severity_data.iterrows():
    ai_application = row['AI Application']
    severity_counts = [int(x) for x in row['Severity (Critical, High, Medium, Low)'].strip('()').split(',')]

    # Add labels with counts to the data points on the radar chart with enhanced visuals
    ax.fill(angles, severity_counts, alpha=0.3)
    ax.plot(angles, severity_counts, label=ai_application)

# Display a legend indicating application names
ax.legend(loc='upper right', bbox_to_anchor=(1.2, 1))

# Show the radar plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the data from the Excel file
df = pd.read_excel('AI data synk.xlsx', engine='openpyxl')

# Extract the relevant columns
applications = df['AI Application']
severity_data = df['Severity (Critical, High, Medium, Low)'].str.strip('()').str.replace(' ', '').str.split(',')

# Convert the severity data into a list
severity_data_list = [[int(val) for val in severity] for severity in severity_data]

# Create a DataFrame for the severity data
severity_df = pd.DataFrame(severity_data_list, columns=['Critical', 'High', 'Medium', 'Low'])

# Create a visually appealing heatmap for the severity distribution
plt.figure(figsize=(10, 6))
ax = sns.heatmap(severity_df, annot=True, fmt='d', cmap='YlGnBu', xticklabels=1, cbar_kws={'label': 'Severity Counts'}, linewidths=0.5, linecolor='gray')

# Set y-axis labels as application names
ax.set_yticklabels(applications, rotation=0, fontsize=10)

# Set labels and title
plt.xlabel('Severity Categories', fontsize=12)
plt.ylabel('AI Application', fontsize=12)
plt.title('Severity Distribution in AI Applications', fontsize=14)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the heatmap plot
plt.tight_layout()
plt.show()
```
